```python
dv.columns = dv.columns.str.replace('Nome órgão solicitante','orgao_solicitante').str.replace('Nome','nome').str.replace('Cargo','cargo').str.replace('Função','funcao').str.replace('Período - Data de início','data_ini').str.replace('Período - Data de fim','data_fim').str.replace('Motivo','motivo').str.replace('Valor diárias','valor_diarias').str.replace('Valor passagens','valor_passagens').str.replace('Valor devolução','valor_devolucao').str.replace('Valor outros gastos','valor_outros_ gastos')
# dv.head()
viagem = dv[['id_viagem', 'orgao_solicitante', 'nome', 'cargo','funcao','data_ini','data_fim', 'motivo','valor_diarias','valor_passagens','valor_devolucao','valor_outros_ gastos']]
viagem.head()
```

```python
dt.columns = dt.columns.str.replace('Sequência Trecho','sequencia').str.replace('Origem - Data','origem_data').str.replace('Origem - País','origem_pais').str.replace('Origem - UF','origem_uf').str.replace('Origem - Cidade', 'origem_cidade').str.replace('Destino - Data','destino_data').str.replace('Destino - País','destino_pais').str.replace('Destino - UF','destino_uf').str.replace('Destino - Cidade', 'destino_cidade').str.replace('Meio de transporte','meio_transporte').str.replace('Número Diárias','numero_diarias').str.replace('Missao','missao').str.replace('?','')
# dt.head()
list_drop_columns = ['Número da Proposta (PCDP)']
trecho = dt.drop(columns=list_drop_columns)

trecho.head()
```
