```python
# lê o arquivo CSV - Viagem
df = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/Trabalho1Bim/viagens/Viagem_limpo.csv')

# mostra o conteúdo do DataFrame
df
```

```python
# detalhes das colunas
df.info()
```

```python
# Convertendo datas

df['data_ini'] = pd.to_datetime(df['data_ini'], format='%d/%m/%Y')
df['data_fim'] = pd.to_datetime(df['data_fim'], format='%d/%m/%Y')

df.info()
# df[df.index.values == 509] #pesquisando por indice
# df
```

```python
# Excluindo linhas com valor total = 0.00
df = df[df['valor_total'] != 0.00]
df.info()
```

```python
# Agrupando cargo por qde procssos
df.groupby(['cargo']).size().reset_index(name='count').sort_values('count', ascending=False)
```

```python
df_motorista = df["cargo"].where(df['cargo'] == "PILOTO")
df_motorista

# print(df['Comedy_Score'].where(df['Rating_Score'] < 50))
```

```python
# Agrupar processos por orgao solicitante
df.groupby(['orgao_solicitante']).size().reset_index(name='count').sort_values('count', ascending=False)
```

```python
# Contagem por orgão com situação de sigilo
df_sigilo = df[df['funcao'] == 'Sigilosa']
df_sigilo.groupby(['orgao_solicitante']).size().reset_index(name='count').sort_values('count', ascending=False)
```

```python
# Classificando orgao pelos gastos
gasto_orgao = df.groupby(['orgao_solicitante', 'valor_total']).size().reset_index(name='count').sort_values('valor_total', ascending=False)
gasto_orgao
```

```python
# Estatistica de dias dos processos de viagens
df_day = df['data_fim'].dt.day - df['data_ini'].dt.day
df_month = df['data_fim'].dt.month - df['data_ini'].dt.month
df_periodo = df_month * 30 + df_day
df_periodo.describe().round(2)
```

```python
# df.info()
# Gastos de Viagens por mes
gastos_mes = df.groupby(df['data_ini'].dt.month)['valor_total'].sum().reset_index(name='total')
gastos_mes
```
