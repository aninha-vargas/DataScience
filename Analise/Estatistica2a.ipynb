```python
# Função de Probabilidade
def probab (A, E):
  resultado = (A / E)*100 
  print('{:.2f}'.format(resultado))
```

```python
# Qual a probabilidade de um processo ser do Inss?
probab (len(df_inss), len(df_aux))
```

```python
# Qual a probabilidade de um processo ser da Policia Federal?
probab (len(df_pf), len(df_aux))
```

```python
# Função com a probailidade de não ocorrer um evento
def probab_nao (A, E):
  resultado = (1- (A / E))*100 
  print('{:.2f}'.format(resultado))
```

```python
# Qual a probabilidade de um processo não ser do Ministério do Trabalho e Emprego?
probab_nao (len(df_mte), len(df_aux))
```

```python
import scipy.stats as stats
stats.probplot(df_aux['valor_total'], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")
plt.show()
```

Pelo teste QQPlot constatado que a amostra acima é não normal.

TESTE MANN WHITNEY

Comparação das medianas do valor total dos Órgãos INSS e Policia Federal.

Ho = NÃO HÁ DIFERENÇA ENTRE AS MEDIANAS DOS VALORES TOTAIS ENTRE INSS E POLICIA FEDERAL : p > 0.05

Ha = HÁ DIFERENÇA ENTRE AS MEDIANAS DOS VALORES TOTAIS ENTRE INSS E POLICIA FEDERAL : p <= 0.05

```python
stat, p = stats.mannwhitneyu(df_inss.valor_total, df_pf.valor_total)
print('Estatística de teste: {}'.format(stat.round(2)))
print('p-valor: {}'.format(p))
```
