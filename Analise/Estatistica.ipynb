```python
# lê o arquivo CSV - Viagem
df = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/Trabalho1Bim/viagens/Viagem_limpo.csv')

# mostra o conteúdo do DataFrame
df
```

```python
# detalhes das colunas
df.info()
```

```python
# Convertendo datas

df['data_ini'] = pd.to_datetime(df['data_ini'], format='%d/%m/%Y')
df['data_fim'] = pd.to_datetime(df['data_fim'], format='%d/%m/%Y')

df.info()
# df[df.index.values == 509] #pesquisando por indice
# df
```

```python
# Excluindo linhas com valor total = 0.00
df = df[df['valor_total'] != 0.00]
df.info()
```

```python
# Agrupar processos por orgao solicitante
df_orgao = df.groupby(['orgao_solicitante']).size().reset_index(name='count').sort_values('count', ascending=False)
df_orgao = df_orgao.head(10)
df_orgao
```

```python
# moda de orgão solicitante
df['orgao_solicitante'].mode()
```

Verificando que o orgão com maior volume de processos de viagem, é o Departamento de Polícia Federal

```python
df_pf = df.loc[df.orgao_solicitante == 'Departamento de Polícia Federal']
df_pf
```

```python
plt.hist(df_pf.cargo)
plt.title('Distribuição de Cargos - Departamento de Polícia Federal')
plt.show()
```

Todos os processos do Departamento de Polícia Federal, são sigilosos.

```python
# Estatistica de dias dos processos de viagens
df_day = df['data_fim'].dt.day - df['data_ini'].dt.day
df_month = df['data_fim'].dt.month - df['data_ini'].dt.month
df_periodo = df_month * 30 + df_day
df_periodo.describe().round(2)
# df_periodo
```

```python
df_periodo.median().round(2)
```

A média de dias dos processos foi de 4,5. Porem a mediana foi de 2, sendo ela o ponto central. Observado uma tendencia de assimetria na distribuiçao.

```python
df['mes'] = df['data_ini'].dt.month
df['mes'].mode()
```

Observado a quantidade de processos que foram distribuídos por mes, sendo novembro o mes que mais teve processos iniciados.

```python
plt.hist(df.mes, 12)
plt.title('Frequencia de processos por mes')
plt.show()
```
