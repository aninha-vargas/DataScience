```python
# Convertendo datas

df['data_ini'] = pd.to_datetime(df['data_ini'], format='%d/%m/%Y')
df['data_fim'] = pd.to_datetime(df['data_fim'], format='%d/%m/%Y')

df.info()
# df[df.index.values == 509] #pesquisando por indice
# df
```

```python
# Excluindo linhas com valor total = 0.00
df = df[df['valor_total'] != 0.00]
df.info()
```

```python
df.describe().round(2)
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

ax = sns.boxplot(x=df["valor_total"])
```

Baseado no valor do 3º quartil e observando estatísticas anteriores, considerado para o estudo processos com valor total maximo de 1.600 que corresponde a 75%. Os demais valores foram desconsiderados, pois foram tratados como outliers.
